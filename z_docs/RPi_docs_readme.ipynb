## 1.2 Address map 
### 1.2.3 ARM physical addresses
- The bus addresses for peripherals are set up to map onto the peripheral bus address range starting at 0x7E000000. Thus a peripheral advertised here at bus address 0x7Ennnnnn is available at physical address 0x20nnnnnn.
- 可以透過  Broadcom BCM2711 ARM Peripherals 文檔來看到 peripherals 對應的 register, 
- 不過有點混亂的地方是文檔中的 peripheral base address 是 0x7Exxxxxx
- 在Raspberry pi 3B+ 會被映射到 0x3Fxxxxxx
  - 所以在 base.h 我們定義 :  MMIO_BASE 0x3F000000UL

- 0x7E215068~0x7E2150C4 is SPI0 and SPI1 

- 每個 addr 對應一個 byte, 所以幾乎都是跳 4 個 byte, = 32bits

- p11 baud rate :  
  $$
  \text{baudrate} = \frac{\text{system\_clock\_freq}}{8\times(\text{baudrate\_reg+1})}
  $$
  - sys_clock_fre = 250 M Hz
  - 設定 mu_baud_rate_reg = 270, 則
   baudrate=   baudrate = 115200,  
- p12-14 datasheet 其實寫錯了, [datasheet error](https://elinux.org/BCM2835_datasheet_errata#p12)
- p12 Title should be AUX_MU_IER_REG. The name in Synopsis is correct
  - These are R/W bits not read only
  - Bits 3:2 are marked as don't care, but are actually required in order to receive interrupts.
  - Bits 1:0 are swaped. bit 0 is receive interrupt and bit 1 is transmit. 
- p13: AUX_MU_ier: The title should be AUX_MU_IIR_REG. The name in the Synopsis is correct 
- p14: 寫錯了 AUX_MU_lcr: 5:2 reserve 0, 
  - [1:0]-> 00->:bit mode , 11:8bit mode 所以要設定成 3才是 8bit mode
- p16: AUX_MU_cntrl[7:0]-> 1:Tx enable, 0:Rx enable
