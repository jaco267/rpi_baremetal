## D13 AArch64 system register descriptions   
### D13.2 General system control register
- D13.2.47 **HCR_EL2**
  - boot.s/sysregs.h
  - HCR_RW: bit31, 讓 EL1 用 AArch64 (64-bit) 執行。
  - The Execution state for EL1 is AArch64. The Execution state for EL0 is determined by the current value of PSTATE.nRW when executing at EL0.
- D13.2.112 **SCR_EL3**, secure config register
  - boot.s
  - 主要和 exception level 相關 
  -  下一層 EL 用 aarch64
  -  Exception levels lower than EL3 are in Non-secure state
  -  and so memory accesses from those Exception levels cannot access Secure memory.

- D13.2.113 **SCTLR_EL1**, system control register (EL1)
  - sysregs.s
  - 可以設定 MMU disable, little_endian 等等
  - 有一個 table 
    - 63....2=C, 1=A, 0=M
    - 每個 flag 會標示 reserve 0 or 1   
    - 大部分 res0, res1 為 bit 29,28,23,22,20,11
- D13.2.137 VBAR_EL1
  - 在 irq.s
  - 用來 設定 EL1 的 exception vector table 位址
  - 告訴 CPU 「發生 exception 時要跳去哪裡」
  - Holds the vector base address for any exception that is taken to EL1.
