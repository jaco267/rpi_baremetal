- armstub 從 EL3 開機   
- boot.S 設定系統暫存器 降到 EL1, 初始化環境, 呼叫 main.c 的 main() 


```s
    // bit:0           bit:2         bit:12        bit:25
    // MMU: bit0 off, D cache: off, I cache: off, little endian  
    ldr x0, =SCTLR_VALUE_MMU_DISABLED
    msr sctlr_el1, x0

    //Hypervisor Control Register //HCR_RW = 1: EL1 使用 AArch64
    ldr x0, =HCR_VALUE
    msr hcr_el2, x0

    // secure config register // RW = 1   → 下一層 EL 用 AArch64
    // NS = 1   → EL1/EL0 是 Non-secure
    ldr x0, =SCR_VALUE
    msr scr_el3, x0
    
    //SPSR(Saved Program Status Register), eret 時 CPU 要變成什麼模式
    //設定spsr_el3, 跳到 EL1  // 使用 SP_EL1h
    ldr x0, =SPSR_VALUE
    msr spsr_el3, x0

    // 設定返回位址
    adr x0, el1_entry
    msr elr_el3, x0
    eret  //跳到 el1
el1_entry:
    adr x0, bss_begin
    adr x1, bss_end
    sub x1, x1, x0
```
