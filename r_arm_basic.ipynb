- ARM instructions  

```
- <operator> <dst>, <src>
- <operator> <dst>, <immediate>
- <operator> <dst>, [address]
```

- Example 
  - mov x0, #0x4  
  - This moves the immediate 4 into the register x0

#### advance instruction 

|instruction      |description                     |example     |
|-----------------|--------------------------------|------------|
|mrs <Xt> sys_reg |move register from system register(read)|mrs x0,CurrentEL|
|msr sys_reg, <Xt>|move to system register (write)         |msr sctrl_el1,x0|
|str val, reg     |store reg, ex xzr=zero register 永遠是0|str xzr,[x0]|

#### system register  
- 不是一般的 x0~x30, 是 CPU 控制用暫存器
- CurrentEL:目前 exception level
- SPSR_EL1 : exception return PC
- ELR_EL1  : exception return 狀態
- VBAR_EL1 : exception return 位置
- DAIF : interrupt mask
- TTBR0_EL1: page table base
- SCTLR_EL1 : MMU / cache 開關





#### system call  
- swi 0 
- "software interrupt" 0x0  
- upon this instruction, the kernel takes an action  
- r7 determines what we do  
- r0-r4 determines how we do it 

- you can search for [arm32 sys call table](https://chromium.googlesource.com/chromiumos/docs/+/master/constants/syscalls.md#arm-32_bit_EABI)

|NR| syscall name  |r7  |arg0(r0)      |arg1(r1) |arg2(r2)    |
|--|---------------|----|--------------|---------|------------|
|0 |restart_syscall|0x00|int error_code|         |            |
|1 |exit           |0x01|              |         |            |
|2 |fork           |0x02|              |         |            |
|3 |read           |0x03|uint fd       |char* buf|size_t count|
|4 |write          |0x04|uint fd       |char* buf|size_t count|
